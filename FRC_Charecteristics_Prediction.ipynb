```python
import pandas as pd
import numpy as np
from scipy.interpolate import interp1d

# Load the CSV files into data frames
graph1 = pd.read_csv('/content/sample_data/SF_10_N_Analysis.csv', header=None)
graph2 = pd.read_csv('/content/sample_data/SF_10_U_Analysis .csv', header=None)

# Drop the first row
graph1 = graph1.drop(graph1.index[0])

# Save the modified DataFrame back to a CSV file
graph1.to_csv('your_file_without_first_row.csv', index=False)

# Drop the first row
graph2 = graph2.drop(graph2.index[0])

# Save the modified DataFrame back to a CSV file
graph2.to_csv('your_file_without_first_row.csv', index=False)

# Convert the second column to a numeric data type
graph1[1] = graph1[1].astype(float)
graph2[1] = graph2[1].astype(float)

# Find the peaks in both graphs
peak_index1 = graph1[1].idxmax()
peak_index2 = graph2[1].idxmax()

# Determine which graph has fewer rows and interpolate
if graph1.shape[0] < graph2.shape[0]:
    # Interpolate graph1
    x = graph1.iloc[peak_index1-10:peak_index1+10, 0].astype(float) # convert string values to float
    y = graph1.iloc[peak_index1-10:peak_index1+10, 1]
    f = interp1d(x, y, kind='linear')
    xnew = pd.Series(np.linspace(x.min(), x.max(), graph2.shape[0]))
    graph1 = pd.concat([xnew, pd.Series(f(xnew))], axis=1)
    graph1.columns = [0, 1]
else:
    # Interpolate graph2
    x = graph2.iloc[peak_index2-10:peak_index2+10, 0].astype(float) # convert string values to float
    y = graph2.iloc[peak_index2-10:peak_index2+10, 1]
    f = interp1d(x, y, kind='linear')
    xnew = pd.Series(np.linspace(x.min(), x.max(), graph1.shape[0]))
    graph2 = pd.concat([xnew, pd.Series(f(xnew))], axis=1)
    graph2.columns = [0, 1]

# Compare the two graphs
mse = ((graph1[1] - graph2[1])**2).mean() # mean squared error
corr = graph1[1].corr(graph2[1], method='pearson') # Pearson correlation coefficient

print(f"Mean squared error: {mse}")
print(f"Pearson correlation coefficient: {corr}")

num_rows = graph2.shape[0]
print(num_rows)


# add first row
# add to end of already fixed data
```
